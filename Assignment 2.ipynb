
# Problem 4 (3 points)

Create a program that prompts for your weight, gender, number of drinks, the amount of alcohol by volume of the drinks consumed, and the amount of time since your last drink. Calculate your blood alcohol content (BAC) using this formula $BAC = (A \times 5.14 / W \times r) - 0.015 \times H$

where

- A is total alcohol consumed, in ounces (oz).
- W is body weight in pounds.
- r is the alcohol distribution ratio:
  - 0.73 for men
  - 0.66 for women
- H is number of hours since the last drink.

Display whether or not it’s legal to drive by comparing the blood alcohol content to 0.08.

## Example Output
<pre>
Your BAC is 0.08
It is not legal for you to drive.
</pre>

## Constraint

- Prevent the user from entering non-numeric values.

```python
def problem4():

    # Weight input
    while True:
        n = input("Enter weight in lbs: ")
        try:
            weight = float(n)
        except ValueError:
            print("Please enter a number.")
            continue
        else:
            if weight < 30:
                print("Enter a valid weight.")
                continue
            else:
                break
    
    # Gender input
    while True:
        n = input("Enter gender (1 for Female / 2 for Male): ")
        try:
            gender = int(n)
        except ValueError:
            print("Please enter a number.")
            continue
        else:
            if gender not in (1,2):
                print("Enter a valid input.")
                continue
            else:
                gender -= 1
                break
    
    # Number of drinks input    
    while True:
        n = input("Enter number of drinks in ounces (oz): ")
        try:
            drinks = float(n)
        except ValueError:
            print("Please enter a number.")
            continue
        else:
            if drinks < 0:
                print("Enter a non-negative number.")
                continue
            else:
                break
    
    # Amount of alcohol by volume of the drinks consumed input    
    while True:
        n = input("Enter amount of alcohol by volume of the drinks consumed as 0-100: ")
        try:
            abv = float(n)
        except ValueError:
            print("Please enter a number.")
            continue
        else:
            if abv < 0:
                print("Enter a non-negative number.")
                continue
            elif abv > 100:
                print("Enter a vaid number.")
                continue
            else:
                abv = abv / 100
                break
    
    # Hours since last drink input    
    while True:
        n = input("Enter the amount of time since the last drink in hours: ")
        try:
            hours_drink = float(n)
        except ValueError:
            print("Please enter a number.")
            continue
        else:
            if hours_drink < 0:
                print("Enter a non-negative number.")
                continue
            else:
                break
    
    # Alcohol distribution ratios for women and men
    r_bac = (0.66, 0.73)

 
    # BAC calculation
    BAC= ((drinks * abv) * (5.14 / weight) * r_bac[gender]) - (0.015 * hours_drink)
    
    legal_limit = 0.08
    
    print("\nYour BAC is {:.2f}".format(BAC))
    
    if BAC >= legal_limit:
        print("It is not legal for you to drive")
    else:
        print("It is legal for you to drive.")
        
    
problem4()
```

The formula for this is

$n = -\frac{1}{30} \times \frac{\log(1 + \frac{b}{p}(1 - (1 + i)^{30}))}{\log(1+i)}$

where

- `i` is the daily rate (APR divided by 365).

- Round fractions of a cent up to the next cent.

```python
def problem11():
    
    import math
    
    # User inputs and data validation
    while True:
        try:
            inp_balance = float(input("What is your balance? "))
            inp_APR = float(input("What is the APR on the card (as a percent)? "))
            inp_monthpay = float(input("What is the monthly payment you can make? "))
        except ValueError:
            print("Please enter a valid number.")
            continue
        else:
            if inp_balance < 0 or inp_APR <  0  or inp_monthpay < 0 or inp_APR > 100:
                print("Please enter a valid number.")
                continue
            else:
                break
    
    
    def calculateMonthsUntilPaidOff(bal, apr, m_pay):
        """
        This function takes a balance, APR, and monthly payments and calculates months until paid off.
        """
        return -(1/30.0) * ((math.log(1+ (bal / m_pay) * (1-(1+(apr/36500))**30))) / (math.log(1+ (apr/36500))))
        
    print("It will take you {:.0f} months to pay off this card.".format(math.ceil(calculateMonthsUntilPaidOff(inp_balance, inp_APR, inp_monthpay))))
    
            
problem11()
```

- Create a function for each type of validation you need to write. Then create a `validateInput` function that takes in all of the input data and invokes the specific validation functions.

```python
def problem12():
    def validateInput(fname, lname, zipcode, id):

        def validate_fname(fname):
            """
            Validates first name against set criteria.
            """
            if len(fname) == 0:
                return "The first name must be filled in."
            elif len(fname) < 2:
                return "{} is not a valid first name. It is too short.".format(fname)
            else:
                return 1
    
        def validate_lname(lname):
            """
            Validates last name against set criteria.
            """
            if len(lname) == 0:
                return "The last name must be filled in."
            elif len(lname) < 2:
                return "{} is not a valid first name. It is too short.".format(lname)
            else:
                return 1

        def validate_zip(zipcode):
            """
            Validates zip code against set criteria.
            """
            if len(zipcode) == 0:
                return "The zip code must be filled in."
            elif len(zipcode) < 5:
                return "{} is not a valid zip code. It is too short.".format(zipcode)
            elif len(zipcode) > 5:
                return "{} is not a valid zip code. It is too long.".format(zipcode)
            else:
                try:
                    i = int(zipcode)
                except ValueError:
                    return "The ZIP code must be numeric."
                else:
                    return 1
        
        def validate_emp_id(id):
            """
            Validates employee ID against set criteria.
            """
            id_list = list(id)
            
            if len(id_list) != 7:
                return "{} is not a valid ID.".format(id)
            
            # BReaks down ID into seperate components
            id_alpha = id_list[:2]
            id_dash = id_list[2]
            id_num = id_list[3:]
            
            id_alpha = "".join(str(e) for e in id_alpha)
            id_num = "".join(str(f) for f in id_num)
            
            if id_alpha.isalpha() and id_num.isdigit() and "-" in id_dash:
                return 1
            else:
                return "{} is not a valid ID.".format(id)
        
        # Creates output string for invalid entries
        s = []
        if validate_fname(fname) != 1: 
            s.append(validate_fname(fname))
        
        if validate_lname(lname) != 1:
            s.append(validate_lname(lname))
        
        if validate_zip(zipcode) != 1:
            s.append(validate_zip(zipcode))
        
        if validate_emp_id(id) != 1:
            s.append(validate_emp_id(id))
        
        if len(s) == 0:
            return "There were no errors found."
        else:
            return '\n'.join(s)
        
        
    # User inputs
    user_fname = input("Enter the first name: ")
    user_lname = input("Enter the last name: ")
    user_zip = input("Enter the ZIP code: ")
    user_empid = input("Enter an employee ID: ")
    
    print(validateInput(user_fname, user_lname, user_zip, user_empid))
        

problem12()
```

# Problem 15 (3 points)

Create a program that generates multiplication tables for the numbers 0 through 12.

## Example Output

<pre>
       1    2  ...   12
  1    1    2        12
  2    2    4        24
  :
 12   12   24       144
</pre>

## Constraint

- Use a nested loop to complete this program.

```python
def problem15():
    
    # Define multiplication table and the spaces for printing each result
    mult_table = 12
    char_space = len(str(mult_table ** 2)) + 1
    
    # Creates the multiplication table
    for i in range(0, mult_table + 1):
        for j in range(0, mult_table + 1):
            if i == 0 and j == 0:
                print('{:<{x}}'.format(" ", x = char_space), end=" ")
            elif j == 0:
                print('{:<{x}}'.format(i, x = char_space), end=" ")
            elif i == 0:
                print('{:<{x}}'.format(j, x = char_space), end=" ")
            else:
                print('{:<{x}}'.format(i * j, x = char_space), end=" ")
        print('\n')
        
problem15()
```

- Ensure that the most used word is at the top of the report and the least used words are at the bottom.

```python
def problem30():
    
    #http://codereview.stackexchange.com/questions/103773/count-the-frequency-of-words-in-a-text-file
    
    from collections import defaultdict
    from collections import OrderedDict
    
    maxLength = 0
    
    file=open("Problem30_Ass2.txt","r")

    # Creates defaultdict initiallized with integers.
    wordcount = defaultdict(int)

    # Increments dictionary values by one if word is found.  Word is dictionary keyword.
    for word in file.read().split():
        wordcount[word] += 1
        maxLength = max(maxLength, len(word))
    
    # Creates an ordered dictionary to store results by most common word.
    wordcount = OrderedDict(sorted(wordcount.items()))

    # Prints the word and a histogram of "*".
    for k,v in wordcount.items():
        print("{:<{x}}".format(k + ": ", x = maxLength + 2) + "*" * v)
    
    
problem30()
```
